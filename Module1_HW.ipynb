We have no null values, the "unknown" strings in categorical variables may be placeholders for missing data. I will try to replace "unknown" with the most frequent category. Mode represents the most common and popular category within that feature. This is most straightforward method for categorical features, unknown entry is most likely to be similar to the most common EXISTING entries, so that why I have chosen the MODE

```python
for column in categorical_features:
    mode = data[column].mode()[0]
    data[column] = data[column].replace('unknown', mode)
```

```python
data.head()
```

Now lets work with outliers

```python
for column in numerical_features:
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_cap = Q1 - 1.5 * IQR
    upper_cap = Q3 + 1.5 * IQR

    # Cap the data only within Q1 and Q3
    data[column] = np.where(data[column] < lower_cap, lower_cap, data[column])
    data[column] = np.where(data[column] > upper_cap, upper_cap, data[column])

capped_descriptive_stats = data[numerical_features].describe()
capped_descriptive_stats
```

Above:
Outlier detection method is interquartile range and we capped them within 1.5*IQR of the Q1 and Q3. SO capping preserves all data points and we do not loose valuable information.I have chosen this method because "Capping means putting a limit on these values". If a value is higher than our upper limit, we bring it down to the upper limit. And similar too low. It helps preventing these extreme values from having an outsized influence on the analysis and focusing on main majority data
